# K Nearest Neighbors

**K-NN** es un algoritmo de aprendizaje supervisado, es decir, requiere que nosotros le suministremos algo de información.

Esta técnica, por tanto, se suele utilizar tras haber aplicado otra técnica como el clustering jerárquico o K-means.

En el siguiente ejemplo, al algoritmo le diremos que existen, de antemano, 2 grupos y a qué grupo pertenece cada uno de esos datos.

El objetivo será que la máquina, al recibir un nuevo dato y teniendo los datos históricos que le hemos suministrado, sea capaz de colocarlo en el grupo correspondiente.

En este caso, serán los "vecinos" los que decidirán a qué grupo pertenece el nuevo caso, basándose en la **distancia**: Los k vecinos más cercanos son los que deciden la clasificación.

El número de grupos, que hay que definir a priori, y **k** será la cantidad de vecinos que decidirán la clasificación. Es decir, los **k vecinos más cercanos**.

En la siguiente imagen vemos que los 3 vecinos más cercanos al punto deciden la clasificación ya que la distancia es pequeña.

```python
Image("resources/knn-dec1.png")
```

En el ejemplo anterior, si dos vecinos hubieran "votado" que el nuevo punto es cruz roja y uno que punto negro, el nuevo dato aún se clasificaría como cruz roja, a pesar de que hubiera un poco menos confianza de que fuera cruz roja.

Por la naturaleza del algoritmo, siempre procuraremos elegir un **valor impar para k**, para, de esta forma, evitar un posible empate a la hora de las votaciones.

Este algoritmo se puede "tunear" penalizando a los puntos que están más lejos del nuevo punto o dando peso al voto de los que están más cerca.
