# Deep Learning y Aprendizaje No Supervisado - Teoría

Una Red Neuronal simula el funcionamiento de las neuronas dentro de un cerebro, combinando construcciones individuales de muchas entradas (columnas en un data set tabular, palabras en un documento word, o pixels en una imagen, en definitiava, datos no estructurados) para predecir un valor de interés. P. ej. si es un precio válido, muy alto o muy bajo, predecir un cancer, etc. 

Este tipo de modelos implica un gran número de parámetros, y particularmente un lenguaje propio de las neuronas, que tiene que ver con la biología, para llevar a cabo mapeos de los datos de entrada en una variable de salida. La complejidad de estos modelos reside en ser capaces de calibrar estos parámetros. En definitiva, pueden llegar a tener cientos o miles de parámetros.

Los algoritmos de Deep Learning están compuestos de unidades muy sencillas que son la estructura básica de las redes neuronales y que, generalmente, llevan a cabo regresiones logísticas.

La red neuronal más sencilla, aquella conformada por una operación lineal sencilla o una sola neurona, es conocida como **perceptrón**.

```python
from IPython.display import Image
Image("resources/perceptron.png")
```

En el perceptrón ocurre que, dada una entrada `x`, la función de decisión nos devuelve 1 si la neurona se excita, es decir, si la existencia de un parámetro `w` (una serie de pesos) multiplicado escalarmente por cada una de las `x` supera un umbral `b` (intercept o umbral de decisión).

Esta función de decisión (como los SVM) trabaja los datos separándolos en dos clases dependiendo de si un punto pertenece o no al hiperplano separador dado por `w`.

Si quisiéramos determinar un parámetro `w` óptimo para este perceptrón usando un data set, podríamos llevar a cabo un algoritmo como el siguiente.

```python
Image("resources/percep-alg.png")
```

Este modelo tiene sus limitaciones prácticas ya que, como vimos con los SVM, no todos los datos son linealmente separables por un hiperplano, por lo que no siempre será posible clasificar todos y cada uno de los puntos, provocando que el algoritmo nunca convergiera.
 
De forma adicional, todos los parámetros van a ser actualizados con la misma regla, significando ello que necesitamos aprender un patrón común acerca de la importancia de cada una de las columnas en todos y cada uno de los puntos.

La salida de la función, finalmente, será 0 ó 1, por lo que solo podemos usar el perceptrón para clasificaciones binarias.

Sin embargo, a pesar de estas limitaciones, este modelo ilustra muy bien los rasgos de una red neuronal compleja. Entrenar el algoritmo nos dará los parámetros buenos para el modelo basándonos en el rol de clasificación ajustado por el ratio de aprendizaje $\alpha$ y el patrón puede ir complicándose de muchas formas.

Generalmente el umbral será binario, pero podemos estudiar otras formas de clasificación, estableciendo una función sigmoidea en lugar de lineal, como la siguiente.

Se pueden combinar elementos de este estilo para desarrollar perceptrones o Redes Neuronales, con múltiples unidades o múltiples capas (incluso con capas ocultas o intermedias) que irán llevando a cabo más y más operaciones de forma constante.

El modelo se puede complicar, pero no es más que la repetición de la operación llevada a cabo por el perceptrón una y otra vez.

Las operaciones del perceptrón, por lo general, será alguna de las tres **funciones de activación** que se muestran a continuación.

```python
Image("resources/funcionesactivacion.png")
```

* La primera nos sonará del núcleo de convolución radial que empleamos en el SVM.
* La segunda es la función rectificadora del máximo entre 0 y x cuya versión más regularizada es, aproximadamente, un logaritmo. 
* La tangente hiperbólica, que nos recuerda a la función radial

Si el valor de la evaluación, en cualquiera de los casos, cae por encima de `0`, la probabilidad supera el 50% y es catalogado como positivo o excitado.

A partir de esta base, existen toda una serie de teorías y métodos diferentes para la obtención de los mejores parámetros como hacer propagación, la optimización a partir de las derivadas parciales, como obtener cotas del error o encontrar para cada una de las neuronas un modelo que se ajuste.

En nuestro contexto de introducción a las redes neuronales, todas estas funciones las encontraremos ya implementadas en el framework `TensorFlow`.
